```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4],[5, 6, 7, 8],"go")
plt.title("Graph_1")
plt.xlabel("days")
plt.ylabel("working hours")
```

```python
plt.figure(figsize = (15,5))
plt.plot([1, 2, 3, 4],[5, 6, 7, 8],"b-")
plt.title("Graph_1")
plt.xlabel("days")
plt.ylabel("working hours")
```

```python
x = np.arange(1,5)
y = x**5
plt.plot([1, 2, 3, 4],[5, 6, 7, 8],"b-", x, y, "r+")
plt.title("Graph_1")
plt.xlabel("days")
plt.ylabel("working hours")
```

```python
x = np.arange(1,6,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title("Graph_1")
plt.xlabel("days")
plt.ylabel("working hours")
```

```python
month=["Jan","Feb","Mar","Apr","May","June","July"]
sales=[60,50,55,45,65,45,60]
plt.bar(month,sales,color="cyan")
plt.title("BarGraph")
plt.xlabel("month")
plt.ylabel("sales")
```

```python
month = ["Jan","Feb","Mar","Apr","May","June","July"]
sales = [60,50,55,65,45,40,65]
plt.barh(month,sales,color="gray")
plt.title("BarGraph")
plt.xlabel("month")
plt.ylabel("sales")
```
